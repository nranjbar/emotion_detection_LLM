```python
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

def train_and_evaluate(train_loader, model, optimizer, lr_scheduler, device, num_epochs=3):
    """
    Function to train the model for multiple epochs and evaluate the performance after each epoch.
    Arguments:
    - train_loader: DataLoader for training data.
    - model: The model to be trained.
    - optimizer: The optimizer for model parameters.
    - lr_scheduler: The learning rate scheduler.
    - device: Device to run the model on (CPU or GPU).
    - num_epochs: Number of training epochs (default is 3).
    
    Returns:
    - A list of precision, recall, and F1 scores for each emotion.
    """
    
    # Lists to store results for each epoch
    precision_scores = []
    recall_scores = []
    f1_scores = []
    
    for epoch in range(num_epochs):
        print(f"Training Epoch {epoch+1}/{num_epochs}")
        
        # Set model to training mode
        model.train()
        total_loss = 0
        all_predictions = []
        all_labels = []

        for batch in train_loader:
            batch = {k: v.to(device) for k, v in batch.items()}  # Move batch to device
            outputs = model(**batch)  # Forward pass
            loss = outputs.loss
            total_loss += loss.item()

            # Backpropagation
            loss.backward()
            optimizer.step()
            lr_scheduler.step()
            optimizer.zero_grad()

            logits = outputs.logits
            predictions = torch.sigmoid(logits).detach().cpu().numpy()
            labels = batch["labels"].detach().cpu().numpy()

            predicted_labels = (predictions > 0.5).astype(int)  # Apply threshold to predictions
            all_predictions.append(predicted_labels)
            all_labels.append(labels)

        # Evaluate the model after each epoch
        all_predictions = np.vstack(all_predictions)
        all_labels = np.vstack(all_labels)
        
        # Calculate precision, recall, and F1 for each emotion separately
        precision = precision_score(all_labels, all_predictions, average=None)  # Per class precision
        recall = recall_score(all_labels, all_predictions, average=None)  # Per class recall
        f1 = f1_score(all_labels, all_predictions, average=None)  # Per class F1
        
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)

        # Print per class metrics for each epoch
        for i, emotion in enumerate(["Anger", "Fear", "Joy", "Sadness", "Surprise"]):
            print(f"Epoch {epoch+1} - {emotion}:")
            print(f"  Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1: {f1[i]:.4f}")
    
    return precision_scores, recall_scores, f1_scores
```

```python
def test_model(test_loader, model, device):
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for batch in test_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            logits = outputs.logits
            predictions = torch.sigmoid(logits).cpu().numpy()
            labels = batch["labels"].cpu().numpy()

            # Apply threshold to get binary predictions
            predicted_labels = (predictions > 0.5).astype(int)
            all_predictions.append(predicted_labels)
            all_labels.append(labels)

    # Convert lists to arrays
    all_predictions = np.vstack(all_predictions)
    all_labels = np.vstack(all_labels)

    # Calculate precision, recall, and F1 for each emotion separately
    precision = precision_score(all_labels, all_predictions, average=None)
    recall = recall_score(all_labels, all_predictions, average=None)
    f1 = f1_score(all_labels, all_predictions, average=None)

    # Print per emotion results
    for i, emotion in enumerate(["Anger", "Fear", "Joy", "Sadness", "Surprise"]):
        print(f"Test Results for {emotion}:")
        print(f"  Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1: {f1[i]:.4f}")

    return all_predictions, all_labels
```
